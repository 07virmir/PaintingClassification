```python
import re

def find_artist(input_string):
  match = re.match(r'^[^\d]*', input_string)
  if match:
      result = match.group(0)
      return result[:-1]
  else:
      return None
```

```python
all_artists = set()

for name in augmented_data:
  artist_name = find_artist(name)
  print(artist_name)
  all_artists.add(artist_name)
```

```python
artists_dict = {}
for name in list(os.listdir(augmented_dir)):
  author_name = find_artist(name)
  if author_name in artists_dict:
    artists_dict[author_name].append(name)
  else:
    artists_dict[author_name] = [name]
```

```python
total = 0
for key in artists_dict.keys():
  print(key)
  print(len(artists_dict[key]))
  total += len(artists_dict[key])
print(total)
```

```python
image_count = 0
for name in list(os.listdir(augmented_dir))[2:]:
  author_name = find_artist(name)
  positive_image = random.choice(artists_dict[author_name])
  curr_artists_dict = artists_dict.copy()
  del curr_artists_dict[author_name]
  random_artist = random.choice(list(curr_artists_dict.keys()))
  random_image = random.choice(curr_artists_dict[random_artist])

  target_anchor_dir = augmented_target_dir + "/" + "Anchor" + "/" + str(image_count) + "_" +  name
  target_positive_dir = augmented_target_dir + "/" + "Positive" + "/" + str(image_count) + "_" + positive_image
  target_negative_dir = augmented_target_dir + "/" + "Negative" + "/" + str(image_count) + "_" + random_image
  curr_img_path = augmented_dir + "/" + name
  positive_image_path = augmented_dir + "/" + positive_image
  negative_image_path = augmented_dir  + "/" + random_image

  # print("Current image path: " + curr_img_path)
  # print("Positive image path: " + positive_image_path)
  # print("Negative image path: " + negative_image_path)
  # print("Target anchor image path: " + target_anchor_dir)
  # print("Target Positive path: " + target_positive_dir)
  # print("Target Negative path: " + target_negative_dir)
  # print("-----")
  shutil.copy(curr_img_path, target_anchor_dir)
  shutil.copy(positive_image_path, target_positive_dir)
  shutil.copy(negative_image_path, target_negative_dir)
  image_count += 1
```
